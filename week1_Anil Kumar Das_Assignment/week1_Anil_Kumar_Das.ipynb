**Your Geometric Explanation (4.3):**
> ### Geometric Meaning of Stretching a Vector

A matrix transformation can change a vector's length and direction.

An eigenvector is a special vector whose direction remains unchanged after the transformation. The matrix only scales (stretches or compresses) it.

If:

A v = λ v

then:

- v is an eigenvector
- λ is the eigenvalue

Interpretation:
- If |λ| > 1, the vector is stretched.
- If 0 < |λ| < 1, the vector is compressed.
- If λ < 0, the vector is flipped and scaled.
- If λ = 1, the vector remains unchanged.

Thus, eigenvectors represent directions that are preserved by the matrix transformation, while eigenvalues tell us how much stretching or compression occurs along those directions.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.

4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt =  np.linalg.svd(X, full_matrices=False)  # YOUR CODE HERE

X_reconstructed =U @ np.diag(S) @ Vt # YOUR CODE HERE — reconstruct from U, S, Vt
X_approx        = S[0] * np.outer(U[:, 0], Vt[0, :])# YOUR CODE HERE — rank-1 approximation

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

>
## Relationship Between SVD and PCA

For a data matrix X, Singular Value Decomposition (SVD) factorizes it as:

X = UΣVᵀ

where:

- **U** contains the left singular vectors. These represent the coordinates (scores) of the samples in the transformed space.
- **Σ (S)** contains the singular values, which indicate the importance of each component.
- **Vᵀ** contains the right singular vectors. These define the principal directions of the data.

### Relationship to PCA

When the data matrix is centered (mean subtracted), PCA can be computed using SVD.

- The **rows of Vᵀ** (or equivalently the columns of V) are the **principal components**.
- The **singular values** indicate how much variance is captured by each principal component.
- The transformed coordinates (PCA scores) are related to **UΣ**.

### Directions of Maximum Variance

The matrix **Vᵀ** contains the directions of maximum variance.

- The first row of Vᵀ corresponds to the first principal component (captures the most variance).
- The second row corresponds to the second principal component.
- Each subsequent component captures less variance than the previous one.

### Dimensionality Reduction

Keeping only the largest k singular values and corresponding singular vectors produces a rank-k approximation:

Xₖ = Uₖ Σₖ Vₖᵀ

This retains the most important information while reducing dimensionality and noise.

### Rank-1 Approximation

Using only the largest singular value and its singular vectors gives the best possible rank-1 approximation of X in the least-squares sense. It captures the dominant pattern in the data.





->
Thus, PCA uses the right singular vectors in Vᵀ as the principal component directions and the singular values in Σ to measure their importance.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')  # YOUR CODE HERE
ks_s2 = stats.kstest(s2, 'norm')  # YOUR CODE HERE
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts) # YOUR CODE HERE — adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)  # YOUR CODE HERE — np.diff
adf_diff = adfuller(ts_diff)  # YOUR CODE HERE
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # YOUR CODE HERE
    epsilon = 1e-10

    # Create bins based on expected distribution
    breakpoints = np.linspace(0, 100, bins + 1)
    breakpoints = np.percentile(expected, breakpoints)

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected) + epsilon
    actual_pct = actual_counts / len(actual) + epsilon

    psi = np.sum((actual_pct - expected_pct) *
                 np.log(actual_pct / expected_pct))

    return psi
    pass

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()

ax.hist(train_dist, bins=30, alpha=0.5, density=True, label='Training Distribution')
ax.hist(drift_dist, bins=30, alpha=0.5, density=True, label='Production Distribution')
# YOUR CODE HERE
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** A change in the relationship between input features and the target variable over time, causing the model's predictions to become less accurate.

- **Covariate drift:** A change in the distribution of input features while the relationship between the features and the target variable remains unchanged.

- **PSI < 0.1:** No significant distribution shift; the model is considered stable.

- **PSI 0.1–0.2:** Minor distribution shift; monitor the model and data closely.

- **PSI > 0.2:** Major distribution shift; investigate the cause and consider retraining the model.

- **Retraining trigger example:** Retrain the model if PSI remains above 0.2 for a sustained period or if model performance metrics (e.g., accuracy, F1-score, RMSE) show significant degradation.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free =  (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)  # YOUR CODE HERE

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free # YOUR CODE HERE

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # YOUR CODE HERE
    p_ham = 1 - prior_spam
    p_word = (p_word_given_spam * prior_spam) + \
             (p_word_given_ham * p_ham)

    posterior = (p_word_given_spam * prior_spam) / p_word
    return posterior
    pass

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means =  np.array([
    np.mean(np.random.choice(population, size=sample_size))
    for _ in range(n_samples)
])
 # YOUR CODE HERE — array of 5000 sample means

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size) # YOUR CODE HERE — predicted std of sample means

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
# YOUR CODE HERE — histogram of sample means + theoretical normal overlay
# Histogram of sample means
ax.hist(sample_means, bins=40, density=True, alpha=0.6, label='Sample Means')

# Theoretical normal distribution from CLT
x = np.linspace(sample_means.min(), sample_means.max(), 500)
ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

### CLT Reflection (6.4)

The Central Limit Theorem (CLT) states that the distribution of sample means becomes approximately normal as the sample size increases, even if the original population is not normally distributed. This is important because it allows statisticians and machine learning practitioners to make reliable inferences about population parameters using sample data.

The CLT justifies the normality assumption used in many classical statistical methods and hypothesis tests, such as t-tests, confidence intervals, and linear regression inference, especially when the sample size is sufficiently large.
