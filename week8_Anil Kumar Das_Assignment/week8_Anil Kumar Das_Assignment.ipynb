# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
import logging

logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')
logger = logging.getLogger(__name__)

# 🤖 AGENT FUNCTION (IMPLEMENTED & IMPROVED)
def agent(query: str):
    logger.info(f"Received query: '{query}'")
    query_lower = query.lower()

    if 'calculate' in query_lower:
        logger.info("Routing to Calculator Tool")
        expr = query_lower.replace('calculate', '').strip()
        return {
            'type': 'calculation',
            'result': calculator(expr)
        }
    elif 'keywords' in query_lower:
        logger.info("Routing to Keyword Extractor Tool")
        text = query_lower.replace('extract keywords from', '').replace('keywords', '').strip()
        return {
            'type': 'keywords',
            'result': extract_keywords(text if text else query)
        }
    else:
        logger.info("Routing to General Response")
        return {
            'type': 'general',
            'result': 'I am a simple smart assistant. I can help with math or extracting keywords.'
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````

````python
import json

# 📁 File Input & Output Implementation
def process_queries_from_file(input_file: str, output_file: str):
    results = []
    try:
        with open(input_file, 'r', encoding='utf-8') as f:
            queries = f.readlines()
        
        for q in queries:
            q = q.strip()
            if q:
                results.append({
                    'query': q,
                    'response': agent(q)
                })
        
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(results, f, indent=4)
            
        print(f"Successfully processed {len(results)} queries. Output saved to {output_file}")
    except FileNotFoundError:
        print(f"Error: The file {input_file} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage (uncomment to test):
# with open('input.txt', 'w') as f:
#     f.write("Calculate 20 * 5\nWhat is machine learning?\nExtract keywords from AI is great")
# process_queries_from_file('input.txt', 'output.json')
````
